```python
import requests
import os

# Base URL of your Flask API
BASE_URL = 'http://192.168.2.46:5000/files'

def list_csv_files():
    """Fetch the list of available CSV files from the API."""
    response = requests.get(BASE_URL)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return []

def download_csv(filename, save_dir='csv_downloads'):
    """Download a specific CSV file from the API."""
    os.makedirs(save_dir, exist_ok=True)
    file_url = f"{BASE_URL}/{filename}"
    response = requests.get(file_url, stream=True)
    if response.status_code == 200:
        file_path = os.path.join(save_dir, filename)
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print(f"Downloaded: {file_path}")
    else:
        print(f"Failed to download {filename}: {response.status_code}")

# List available CSV files
csv_files = list_csv_files()
print("Available CSV Files:", csv_files)

# Download all CSV files
for file in csv_files:
    download_csv(file)
```
